# Hyperparameter Tunning Using Keras

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import tensorflow as tf 
import keras
from keras.models import Sequential
from keras.layers import  Dense
```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
data = load_diabetes()
df =pd.DataFrame(data.data,columns = data.feature_names)
df['target'] =data.target
```

```python
df
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(df.drop(columns=['target']),df['target'],test_size=0.2,random_state=42)
```

```python
import kerastuner as kt
```

```python

from keras.activations import sigmoid,relu,leaky_relu,tanh
from keras.losses import mean_squared_error, mean_absolute_error , binary_focal_crossentropy,categorical_crossentropy,sparse_categorical_crossentropy
from keras.regularizers import l2
from keras.optimizers import  Adamax,Nadam,Ftrl,SGD,Adam



def build_model(hp):
    
    activation =hp.Choice('activation',values =['sigmoid','relu','tanh','leaky_relu',])
    loss =hp.Choice('loss',values=['mse','mae','binary_crossentropy','catergorical_crossentropy','sparse_categorical_crossentropy'])
    optimizer = hp.Choice('optimizer',values =['Adam','Adamax','Nadam','Ftrl','SGD'])
    
    model1 =Sequential([Dense(150,activation=activation,input_dim=2,kernel_regularizer=l2(0.05) ),
                    Dense(100,activation=activation,kernel_regularizer=l2(0.01)),
                    Dense(50,activation=activation,kernel_regularizer=l2(0.0025)),
                    Dense(1,activation=activation)])

    model1.compile(loss =loss,optimizer = optimizer,metrics =['accuracy'])
    
    return model1
```

```python
tuner = kt.RandomSearch(build_model ,objective='val_accuracy',max_trials= 5)
```

```python
tuner.search(X_train,Y_train,epochs =5,validation_data =(X_test,Y_test))
```
