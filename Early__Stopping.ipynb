using callbacks  of keras



```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras

```

```python

from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from mlxtend.plotting import plot_decision_regions

import warnings
#warnings.filterwarnings("ignore")  # Suppress warnings

from pylab import rcParams
import seaborn as sns
from sklearn.datasets import make_circles  # Fixed incorrect import statement
```

```python

from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X,y = make_circles(n_samples=100 ,noise=0.1,random_state=1)
plt.scatter(X[:,0],X[:,1],c =y)
```

```python
X_train,X_test ,Y_train,Y_test =train_test_split(X,y,test_size=0.2,random_state=1)

model =Sequential()
model.add(Dense(256,input_dim =2,activation='relu'))
model.add(Dense(1,activation="sigmoid"))

model.compile(loss ="binary_crossentropy",optimizer ='adam',metrics =["accuracy"])


history = model.fit(X_train,Y_train,epochs =3000,validation_data=(X_test,Y_test),verbose =0)
```

```python
plt.plot(history.history['loss'],label='training_data')
plt.plot(history.history['val_loss'],label ="testing_data")
```

```python
# Early Stoping
earlystopping = EarlyStopping(monitor='val_loss',
                         patience =300,
                         min_delta =0.0000001,verbose=1,mode ='auto',
                         baseline =None,
                         restore_best_weights =False)



```

```python
X_train,X_test ,Y_train,Y_test =train_test_split(X,y,test_size=0.2,random_state=1)

model =Sequential()
model.add(Dense(256,input_dim =2,activation='relu'))
model.add(Dense(1,activation="sigmoid"))

model.compile(loss ="binary_crossentropy",optimizer ='adam',metrics =["accuracy"])


history = model.fit(X_train,Y_train,epochs =3000,validation_data=(X_test,Y_test),verbose =0,callbacks=earlystopping) # callbacks = EarlyStopping


```

```python
plt.plot(history.history['loss'],label='training_data')
plt.plot(history.history['val_loss'],label ="testing_data")
```
